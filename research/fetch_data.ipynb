```python
import time
import json
from confluent_kafka import Producer
import requests
from bs4 import BeautifulSoup

# Kafka producer configuration
producer_config = {
    'bootstrap.servers': 'localhost:9092',  # Change to your Kafka broker URL
    'client.id': 'train-status-producer',
}

producer = Producer(producer_config)

# Function to fetch train data from the website
def fetch_train_status(train_num):
    URL = "https://rappid.in/apis/train.php?train_no="+train_num  # Replace with the actual running status page URL
    response = requests.get(URL)
    return json.loads(response.text)

# Kafka producer sending train status to Kafka topic
def send_to_kafka(data):
    print(data['train_name'])
    producer.produce('train-status', key=data['train_name'].split(" ")[0], value=json.dumps(data))
    # for train in data:
    #     print(train.split(" ")[0])
        # producer.produce('train-status', key=str(train.split(" ")[0]), value=json.dumps(train))
        # producer.flush()

if __name__ == "__main__":
    while True:
        train_data = fetch_train_status("12135")
        if train_data:
            send_to_kafka(train_data)
        time.sleep(1)  # Scrape every minute
        break
```

```python
from confluent_kafka import Consumer
import json

# Kafka consumer configuration
consumer_config = {
    'bootstrap.servers': 'localhost:9092',
    'group.id': 'train-status-consumer-group',
    'auto.offset.reset': 'earliest'
}

consumer = Consumer(consumer_config)

# Subscribe to the train-status topic
consumer.subscribe(['train-status'])

# Consume messages from Kafka
def consume_messages():
    while True:
        msg = consumer.poll(1.0)  # Timeout of 1 second
        if msg is None:
            continue
        if msg.error():
            print(f"Error: {msg.error()}")
            continue

        # Parse the message (train status data)
        train_data = json.loads(msg.value().decode('utf-8'))
        print(f"Received train data: {train_data}")
        # Add your logic here (e.g., store in database, update UI, etc.)

if __name__ == "__main__":
    try:
        consume_messages()
    except KeyboardInterrupt:
        pass
    finally:
        # Ensure consumer is closed properly
        consumer.close()
```
